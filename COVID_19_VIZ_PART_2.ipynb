```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.dates as mdates
```

```
data = pd.read_csv("covid_19_india.csv")
data.tail()
```

```
data.set_index("Sno",inplace = True)
```

```
data.drop(["Time","ConfirmedIndianNational","ConfirmedForeignNational"],axis = 1,inplace = True)
```

```
data.index.name = ""
#data["Date"] = pd.to_datetime(data["Date"])
#data.sort_values(by = "Date")
```

```
data.head()
```

```
data["Date"] = pd.to_datetime(data["Date"],format = '%d/%m/%y')
data.sort_values(by = "Date")
data.head()
```

```
def state_data(statename,dataset):
  return dataset[dataset['State/UnionTerritory']==statename]
```

```
MPdata = state_data('Madhya Pradesh',data)
```

```
MPdata
```

```
statewise = data.groupby('State/UnionTerritory').max()
```

```
statewise.drop('Date',axis = 1,inplace = True)
```

```
statewise
```

```
def plot_data(data,statename):
  fig,ax = plt.subplots(figsize=(5,4))
  ax.plot(data['Date'],data['Confirmed'])
  ax.set_title(statename)
  return figure
```

```
fig = plt.figure(figsize=(40,90))

for statenum in range(len(statewise.index)):
    state = statewise.index[statenum]
    statedata = state_data(state,data)
    ax = fig.add_subplot(9,4,statenum+1)
    ax.plot(statedata['Date'],statedata['Confirmed'],lw = 5,color = 'purple')
    ax.plot(statedata['Date'],statedata['Cured'],lw = 5,color = 'green')
    ax.plot(statedata['Date'],statedata['Deaths'],lw = 5,color = 'red')
    ax.xaxis.set_major_locator(mdates.WeekdayLocator())
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
    ax.set_title(state,size = 30)
    ax.tick_params(labelsize=15)

plt.show()
fig.savefig('statewisedata.png')
```

```
ax = statewise.plot.bar(y = 'Cured/Confirmed',figsize = (15,7),color = 'green')
for p in ax.patches:
    ax.annotate(str(round(p.get_height(),1)), (p.get_x() * 0.995, p.get_height() + 1.01))
ax.tick_params(labelsize=15)
ax.set_xlabel('State/UT',size = 20)
ax.set_title("Percentage of Patients Cured",size = 30)

plt.savefig('CURED_PERCENTAGE.png',bbox_inches='tight',pad_inches=1)
files.download( "CURED_PERCENTAGE.png" )
```
